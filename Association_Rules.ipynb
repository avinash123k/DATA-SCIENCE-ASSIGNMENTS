 ### Preprocessing (drop NA, duplicates, filter positive qty, keep InvoiceNo & Description)
 The data is already in a single column of comma-separated items, which is handled in the subsequent steps.


```python
# Remove missing values
df_clean = df.dropna()
```

```python
# Convert single column into list of transactions
transactions = df.iloc[:,0].apply(lambda x: x.split(','))

# Convert to list of lists
transactions = transactions.tolist()

print("Total transactions:", len(transactions))
print("First 5 transactions:")
for t in transactions[:5]:
    print(t)
```

###One-Hot Encoding Transactions

```python
# Encode transactions into one-hot matrix
te = TransactionEncoder()
te_array = te.fit(transactions).transform(transactions)

basket = pd.DataFrame(te_array, columns=te.columns_)
print("Basket shape:", basket.shape)
basket.head(10)
```

##Association Rule Mining:

###Apriori Frequent Itemsets

We use the Apriori algorithm to extract frequent itemsets from the dataset. The min_support parameter helps in filtering out less frequent itemsets. Lowering the support value (e.g., 0.01) allows us to capture more relationships, even if they are not very common.

```python
# Find frequent itemsets
freq_items = apriori(basket, min_support=0.02, use_colnames=True)
freq_items['length'] = freq_items['itemsets'].apply(len)

print("Number of frequent itemsets:", len(freq_items))
freq_items.sort_values(by='support', ascending=False).head(10)
```

After obtaining frequent itemsets, we generate association rules using association_rules().

* Support shows how often a rule occurs in the dataset.

* Confidence indicates how often the rule is correct.

* Lift tells us the strength of the association compared to random chance.

By setting min_threshold=1.0 for lift, we only keep rules that have a positive relationship.

```python
# Generate rules
rules = association_rules(freq_items, metric="confidence", min_threshold=0.3)

# Filter by lift > 1.2
rules = rules[rules['lift'] >= 1.2]

print("Number of rules found:", len(rules))
rules[['antecedents','consequents','support','confidence','lift']].head(20)
```

Sorting & Selecting Interesting Rules

We sort the rules by confidence and lift to identify the strongest and most relevant patterns.
This helps us understand which items are frequently purchased together and which products strongly influence others.

```python
# Show top 10 association rules sorted by lift
top_rules = rules.sort_values(by="lift", ascending=False).head(10)
top_rules[["antecedents", "consequents", "support", "confidence", "lift"]]
```

```python
# Apply Association Rule Mining with different thresholds to find varied relationships

# Using slightly lower support and confidence to discover more rules
frequent_itemsets_varied = apriori(basket, min_support=0.01, use_colnames=True)

rules_varied = association_rules(
    frequent_itemsets_varied,
    metric="lift",
    min_threshold=1.0
)

# Sort rules by confidence and lift for better results
rules_varied = rules_varied.sort_values(['confidence', 'lift'], ascending=[False, False])

# Display top 15 rules
rules_varied.head(15)
```

The above code:
* Uses min_support=0.01 (1%) to capture less frequent but still relevant item combinations.

* Keeps lift > 1 to ensure rules are meaningful (positive association).

* Sorts by confidence and lift, so you can clearly see the strongest relationships.

##Interview Questions


1.   What is lift and why is it important in Association rules?
2.	What is support and Confidence. How do you calculate them?
3.	What are some limitations or challenges of Association rules mining?
item



1. **Lift** is a measure of how much more often two items occur together than we would expect if they were independent. A lift value greater than 1 indicates a positive association (the presence of one item increases the likelihood of the other). It’s important because it highlights relationships that are not just frequent but also meaningful, filtering out coincidental co-occurrences.



2. **Support** tells us how frequently an itemset appears in the dataset. For example, if eggs appear in 50 out of 1,000 baskets, the support for eggs is 50/1000 = 0.05 (5%).
   **Confidence** measures the strength of an association rule. It’s the probability of finding item B in transactions that already contain item A. For example, if 30 out of 50 transactions with eggs also include milk, then confidence for eggs → milk is 30/50 = 0.6 (60%).

3. **Limitations of Association Rule Mining** include:

   * It can generate a very large number of rules, many of which may not be useful.
   * Rules may be statistically strong but not practically meaningful.
   * It works best with categorical data but struggles with continuous/numerical values.
   * Rare but important patterns might be missed if minimum support is set too high.
   * Computationally expensive on very large datasets.
